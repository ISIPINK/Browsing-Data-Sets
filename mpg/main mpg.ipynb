```python
from plotnine import *
import pandas as pd
from plotnine.data import mpg
```

```python
print(mpg.describe())
mpg.isnull().sum()
mpg.info()
mpg.duplicated().sum()
mpg.to_csv("mpg.csv",index=False)
```

```python
pp = ( 
    ggplot(mpg)
    +aes(x ="displ",y="cyl",color="class" )
    +geom_jitter()
)
print(pp)
```

```python
pp = ( 
    ggplot(mpg)
    +aes(x ="year",y="cyl",color="class" )
    +geom_jitter()
)
print(pp)
```

```python
print(mpg["year"].value_counts())
# print(mpg["model"].value_counts())
print(mpg["drv"].value_counts())

def proces_trans(x):
    if "manual" in x:
        return "manual"
    return "auto" 
mpg['trans'] = mpg['trans'].apply(proces_trans)
print(mpg["trans"].value_counts())
```
